```python
#library

import pandas as pd
```

```python
## reading data
data=pd.read_csv('Class_List_ATOC184_UPDATED.csv', skiprows=9, encoding = "latin1", header=None)
```

```python
# Correct column names
correct_column_names = ['Record Number', 'Student Name', 'ID', 'Reg Status', 'Level', 'Credits',
                        'Degree', 'Program', 'Major', 'Class', 'Telephone', 'Email', 'Personal Pronouns', 'None']

# Assign the correct column names
data.columns = correct_column_names
```

```python
# set first column as the index
data.set_index('Record Number', inplace=True)
```

```python
# selecting my assigned students
myStudent=data[79:94]
myStudent
```

```python
# selecting their mails
myStudent_email=myStudent[['Email']]
```

```python
## saving their mail to txt

# Define the file path for the txt file
txt_file_path = "output.txt"

# Save the DataFrame to a txt file with tab ('\t') separator
myStudent_email.to_csv(txt_file_path, sep='\t', index=False)
print('Done!')
```
